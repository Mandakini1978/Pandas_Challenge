```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchase_data.columns
```

```python
#Number of Unique Items calculation

length_unique_item = len(purchase_data['Item ID'].unique())

# Average Price calculation

Avg_price = round(purchase_data['Price'].mean(),2)

# Number of Purchases calculation

Number_of_Purchase = (purchase_data['Purchase ID']).count()

# Total Revenue calculation

Total_Revenue = (purchase_data['Price']).sum()

# Summary Dataframe creation and Summary generation

purchase_analysis_summary = [{"Number of Unique Items" : length_unique_item ,"Average Price": Avg_price,"Number of Purchases" : Number_of_Purchase ,"Total Revenue" : Total_Revenue }]

purchase_analysis_summary_df = pd.DataFrame(purchase_analysis_summary)

# Formatting data 

purchase_analysis_summary_df["Total Revenue"] = purchase_analysis_summary_df["Total Revenue"].map("${:.2f}".format)

# Display output

purchase_analysis_summary_df = purchase_analysis_summary_df.set_index(['Number of Unique Items','Average Price','Number of Purchases','Total Revenue'])

purchase_analysis_summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
print(purchase_data["Age"].max())
print(purchase_data["Age"].min())
```

```python
# Create bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]

# Create labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]
```

```python
# Slice the data and place it into bins
pd.cut(purchase_data["Age"], bins, labels=group_labels).head()
```

```python
# Place the data series into a new column inside of the DataFrame
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)
purchase_data.head()
```

```python
#Calculate the numbers and percentages by age group

Total_Count = purchase_data.groupby('Age Group')['SN'].nunique()
total_player = len(purchase_data['SN'].unique())

Percentage = Total_Count/ total_player *100

#Create a summary data frame to hold the results

merge_Player_df = pd.merge(Total_Count, Percentage, on="Age Group", how="outer", suffixes=("Total Count", "Percentage _Of_Players"))

merge_Player_df = merge_df.rename(columns = {'SNTotal Count':'Total Count','SNPercentage _Of_Players':'Percentage_Of_Players' })

#round the percentage column to two decimal points
merge_Player_df["Percentage_Of_Players"] = merge_Player_df["Percentage_Of_Players"].map("{:.2f}%".format)

#Display data
merge_Player_df


```
